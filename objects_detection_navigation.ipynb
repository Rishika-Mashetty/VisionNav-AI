```python
import subprocess
import cv2
import numpy as np
import time
from ultralytics import YOLO
from gtts import gTTS
from IPython.display import Audio, display
from google.colab.patches import cv2_imshow
from skimage.measure import shannon_entropy

model = YOLO("yolov8m.pt")
stream_url = url

def speak(msg):
    print("­Ъћі", msg)
    tts = gTTS(text=msg, lang='en')
    tts.save("alert.mp3")
    display(Audio("alert.mp3", autoplay=True))

ffmpeg_cmd = [
    'ffmpeg', '-hide_banner', '-loglevel', 'error',
    '-i', stream_url,
    '-vf', 'fps=2',
    '-f', 'image2pipe',
    '-pix_fmt', 'bgr24',
    '-vcodec', 'rawvideo',
    '-'
]

w_land, h_land = 640, 480
frame_size = w_land * h_land * 3
pipe = subprocess.Popen(ffmpeg_cmd, stdout=subprocess.PIPE, bufsize=10**8)

last_alert = 0
cooldown = 4
displayed = 0

print("­ЪДа Indoor Navigation + Wall Detection Running...")

try:
    while True:
        raw = pipe.stdout.read(frame_size)
        if not raw:
            print("Рџа№ИЈ No frame received.")
            continue

        frame = np.frombuffer(raw, np.uint8).reshape((h_land, w_land, 3))
        frame = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
        h, w = frame.shape[:2]

        if displayed < 5:
            cv2_imshow(frame)
            displayed += 1

        results = model(frame)[0]
        now = time.time()
        alert_issued = False

        # Object-based alerts (bottom 10% check)
        for box, cls_id in zip(results.boxes.xyxy, results.boxes.cls):
            x1, y1, x2, y2 = box.tolist()
            label = model.names[int(cls_id)]

            if y2 >= h * 0.90:
                cx = (x1 + x2) / 2

                if cx < w * 0.33:
                    msg = f"{label} ahead on the left. Please move right."
                elif cx > w * 0.66:
                    msg = f"{label} ahead on the right. Please move left."
                else:
                    msg = f"{label} ahead. Please stop."

                if now - last_alert > cooldown:
                    speak(msg)
                    last_alert = now
                    alert_issued = True
                break

        # Wall detection (center-bottom homogeneity or crowding)
        if not alert_issued:
            # Count how many boxes overlap the center-bottom
            overlap_count = 0
            for box in results.boxes.xyxy:
                x1, y1, x2, y2 = box.tolist()
                cx = (x1 + x2) / 2
                cy = (y1 + y2) / 2
                if (w * 0.33 < cx < w * 0.66) and (cy > h * 0.75):
                    overlap_count += 1

            # If РЅЦ 2 objects in center-bottom Рєњ wall likely close
            if overlap_count >= 2:
                if now - last_alert > cooldown:
                    speak("Wall ahead. Please stop.")
                    last_alert = now
                    continue

            # If no objects: check flatness using entropy
            if len(results.boxes) == 0:
                gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                center_patch = gray[int(h*0.7):, int(w*0.3):int(w*0.7)]
                entropy = shannon_entropy(center_patch)

                if entropy < 4.0:  # Low texture = possibly a blank wall
                    if now - last_alert > cooldown:
                        speak("Wall ahead. Please stop.")
                        last_alert = now

except KeyboardInterrupt:
    print("­ЪЏЉ Stopped by user.")
finally:
    pipe.terminate()
    print("­ЪЊ┤ Stream closed.")
```
